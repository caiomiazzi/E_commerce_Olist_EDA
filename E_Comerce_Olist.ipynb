```python
# Importando Bibliotecas
import numpy as np 
import pandas as pd
from scipy import stats 
import os
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df_item = pd.read_csv("olist_order_items_dataset.csv")
df_reviews = pd.read_csv("olist_order_reviews_dataset.csv")
df_orders = pd.read_csv("olist_orders_dataset.csv")
df_products = pd.read_csv("olist_products_dataset.csv")
df_geolocation = pd.read_csv("olist_geolocation_dataset.csv")
df_sellers = pd.read_csv("olist_sellers_dataset.csv")
df_order_pay = pd.read_csv("olist_order_payments_dataset.csv")
df_customers = pd.read_csv("olist_customers_dataset.csv")
df_category = pd.read_csv("product_category_name_translation.csv")
```

```python
df_all = df_orders.merge(df_item, on='order_id', how='left')
df_all = df_all.merge(df_order_pay, on='order_id', how='outer', validate='m:m')
df_all = df_all.merge(df_reviews, on='order_id', how='outer')
df_all = df_all.merge(df_products, on='product_id', how='outer')
df_all = df_all.merge(df_customers, on='customer_id', how='outer')
df_all = df_all.merge(df_sellers, on='seller_id', how='outer')

print(df_all.shape)
```

```python
def resumo_tabela(df):
    print(f"Dataset Shape: {df.shape}")
    sumario = pd.DataFrame(df.dtypes,columns=['dtypes'])
    sumario = sumario.reset_index()
    sumario['Name'] = sumario['index']
    sumario = sumario[['Name','dtypes']]
    sumario['Missing'] = df.isnull().sum().values    
    sumario['Uniques'] = df.nunique().values
    sumario['First Value'] = df.loc[0].values
    sumario['Last Value'] = df.iloc[-1, :].values
    

    for name in sumario['Name'].value_counts().index:
        sumario.loc[sumario['Name'] == name, 'Entropy'] = round(stats.entropy(df[name].value_counts(normalize=True), base=2),2) 

    return sumario
```

```python
import seaborn as sns  # Importa o seaborn
import matplotlib.pyplot as plt
import pandas as pd

# Assume que df_all é o seu DataFrame original contendo a coluna 'price'

# Definindo variáveis
limite_superior_x = 1000
numero_barras = 500

# Cria o histograma com ajustes
ax = sns.displot(data=df_all, x='price', kind='hist', color='blue', bins=numero_barras)

# Verifica o tipo de objeto e define o limite do eixo x
tipo_objeto = type(ax)

if tipo_objeto == sns.FacetGrid:  # Usa o seaborn.FacetGrid
    for sub_ax in ax.axes.flat:
        sub_ax.set_xlim(0, limite_superior_x)
else:
    ax.set_xlim(0, limite_superior_x)

# Adicionando a curva de densidade de probabilidade (KDE)
sns.displot(data=df_all, x='price', kind='kde', color='red', ax=ax)

# Calculando a média e a moda
media = df_all['price'].mean()
moda = df_all['price'].mode()[0]

# Adicionando linhas da média e da moda
plt.axvline(media, color='green', linestyle='dashed', label='Média')
plt.axvline(moda, color='purple', linestyle='dotted', label='Moda')

# Adicionando título, rótulos e legenda
plt.title('Distribuição dos Preços dos Produtos (até R$ 1.000)')
plt.xlabel('Preço em R$')
plt.ylabel('Frequência')
plt.legend()

# Mostra o gráfico
plt.show()

```
